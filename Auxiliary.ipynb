```python
def split_path(data_path,label = 'orig', transform_path =  '/home/chinmay/datatset/train/'):
    import shutil
    folders = sorted(os.listdir(data_path))
    x = 0
    destination = ''
    for image_name in folders:
        #print(image_name)
        if str(image_name).endswith('0.png'):
            destination = transform_path + label + str(x)+"/"
            os.mkdir(destination)
            print("Changing to ",destination)
            x = x+1 
        shutil.copy2(data_path+image_name,destination)
```

### This portion is used for splitting deep fake videos

```python
import os
import shutil
```

```python
def new_dir (base_path, type_name,dir_index):
    dest_dir = base_path+type_name+str(dir_index)+"/"
    os.mkdir(dest_dir)
    return dest_dir
```

```python
def split_function(base_path, dest_base_path, z, type_name = "df_split/", orig_dir_ext = "df"):
    ctr = 0 #A file .DS_Store is present which screws the index. Remove it and make ctr=0
    dir_index = 0
    dest_dir = new_dir(base_path=dest_base_path, type_name=type_name, dir_index=dir_index)
    orig_base_dir = base_path + orig_dir_ext
    for img in z:
        full_file_name = os.path.join(orig_base_dir, img)
        shutil.copy(full_file_name, dest_dir)
        ctr = ctr + 1
        if ctr % 30 == 0:
            dir_index = dir_index + 1
            dest_dir = new_dir(base_path=dest_base_path, type_name=type_name, dir_index=dir_index)
```
